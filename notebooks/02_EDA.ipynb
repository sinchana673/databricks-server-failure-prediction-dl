```python
# EDA for Silver Tables
import pandas as pd
import matplotlib.pyplot as plt

# Load Silver tables
server_events = spark.read.table('workspace.server_failure.server_events_silver').toPandas()
server_logs = spark.read.table('workspace.server_failure.server_logs_silver').toPandas()
server_metrics = spark.read.table('workspace.server_failure.server_metrics_silver').toPandas()

# Events: Count by event_type, basic stats
print(server_events['event_type'].value_counts())
print(server_events.describe(include='all'))
plt.figure(figsize=(8,4))
server_events['event_type'].value_counts().plot(kind='bar', title='Event Type Distribution')
plt.show()

# Logs: Count by log_level, basic stats
print(server_logs['log_level'].value_counts())
print(server_logs.describe(include='all'))
plt.figure(figsize=(8,4))
server_logs['log_level'].value_counts().plot(kind='bar', title='Log Level Distribution')
plt.show()

# Metrics: Summary stats + histograms
print(server_metrics[['cpu','memory','disk']].describe())
server_metrics[['cpu','memory','disk']].hist(bins=15, figsize=(10,4))
plt.suptitle('Server Metrics Distribution')
plt.show()
```
